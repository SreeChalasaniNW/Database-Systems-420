```python
# create data frame with Heavy buyer
con.text_factory = bytes
#create a cursor to query the select statement
cur = con.cursor()
#execute the select query using cursor.execute method
cur.execute("""select acctno, ytd_sales_2009 
from customer where ytd_sales_2009 >
(SELECT
  ytd_sales_2009 AS 'ytd_sales_2009 90%'
FROM customer
WHERE buyer_status='ACTIVE'
ORDER BY ytd_sales_2009 ASC
LIMIT 1
OFFSET (SELECT
         COUNT(*)
        FROM customer
        WHERE buyer_status='ACTIVE') * 9 / 10 - 1) order by 2 desc""")
#prepare the list of heavybuyers
heavyBuyer = list(cur)
#get the accountno from the list using zip .Heavybuyer has 
#2 columns and we want 1st column
acctno_heavyBuyer, ytd_sales_hb = zip(*heavyBuyer)
```

```python
#create a dataframe and prepare the data
df_custSum = pd.DataFrame()
#add acctno,zip,zip4,hh amount to the dataframe
df_custSum['acctno'] = df_customer['acctno']
df_custSum['zip'] = df_customer['zip']
df_custSum['zip4'] = df_customer['zip4']
df_custSum['inc_scs_amt_v4'] = df_customer['inc_scs_amt_v4']
#lamdba function to assign Y to heavybuyers who are above 90%
df_custSum['heavy_Buyer'] = df_customer['acctno'].apply(
    lambda x: 'Y' if x in acctno_heavyBuyer else 'N')
#Add gender to the dataframe
df_custSum['Gender1'] = df_customer['adult1_g']
df_custSum['Gender2'] = df_customer['adult2_g']
```

```python
#b) Add to each customer's record whether the customer has the following credit cards: #AMEX, Discover, VISA, and Mastercard
```

```python
#create a query to display the creditcard details and execute cursor
cur.execute(""" SELECT acctno, 
       CASE WHEN ((amex_prem IS NOT NULL AND amex_prem='Y') or 
       (amex_reg is not null and amex_reg='Y'))
       THEN 'Y'
       ELSE 'N'
       END AS AMEX,
       CASE WHEN ((visa_prem IS NOT NULL AND visa_prem='Y') or 
       (visa_reg is not null and visa_reg='Y'))
       THEN 'Y'
       ELSE 'N'
       END AS VISA,
       CASE WHEN ((mc_prem IS NOT NULL AND mc_prem='Y') or 
       (mc_reg is not null and mc_reg='Y'))
       THEN 'Y'
       ELSE 'N'
       END AS MASTER,
       CASE WHEN ((disc_prem IS NOT NULL AND disc_prem='Y') or 
       (disc_reg is not null and disc_reg='Y'))
       THEN 'Y'
       ELSE 'N'
       END AS DISCOVERY
FROM customer""")
#cur.fetchall() will fetche the rows from the cursor
df_creditCard = pd.DataFrame(cur.fetchall())
#get the column names from cursor using cur.description and add to the datafraem
field_names = [i[0] for i in cur.description]
df_creditCard.columns = field_names
#df_creditCard
```

```python
#Merge creditcard dataframe and custsum to get all the columns
df_custSum_final = pd.merge(df_creditCard,df_custSum, 
                            on='acctno', how='inner')
df_custSum_final
```

```python
#get the customer from the csv
df_customers  = pd.read_csv("customer.csv")
#df_customers
#filter the customer table by buyer_status Active or Lapsed
buyer_status = (df_customers['buyer_status'] == "LAPSED")
# Select all casess where buyer_status is ACTIVE OR LAPSED
df_customerStatus = df_customers[buyer_status]
df_customerStatus1 = df_customerStatus.loc[:,['acctno','buyer_status',
                                              'ltd_sales']]
```

```python
#add a new column to assign a flag to the department description for each customer and assign all Y's
df_item1['Flag'] = 'Y'
#Make the long and narrow dataframe to short and wide by using unstack
df_deptdescr = df_item1.groupby(
    ['acctno', 'deptdescr']).Flag.max().unstack('deptdescr')
#unstack will return 7 columns which has the department description in it.
#assign N where ever the row is not Y
df_deptdescr['Appliances'] = df_deptdescr['Appliances'].apply(
    lambda x: 'Y' if x =='Y' else 'N')
df_deptdescr['Cameras & Camcorder Accessori'] = df_deptdescr[
    'Cameras & Camcorder Accessori'].apply(lambda x: 'Y' if x =='Y' else 'N')
df_deptdescr['Mobile Electronic Accessories'] = df_deptdescr[
    'Mobile Electronic Accessories'].apply(lambda x: 'Y' if x =='Y' else 'N')
df_deptdescr['Home Audio'] = df_deptdescr['Home Audio'].apply(
    lambda x: 'Y' if x =='Y' else 'N')
df_deptdescr['Mobile Electronics'] = df_deptdescr[
    'Mobile Electronics'].apply(lambda x: 'Y' if x =='Y' else 'N')
df_deptdescr['Portable Electronics'] = df_deptdescr[
    'Portable Electronics'].apply(lambda x: 'Y' if x =='Y' else 'N')
df_deptdescr['Small Appliances'] = df_deptdescr[
    'Small Appliances'].apply(lambda x: 'Y' if x =='Y' else 'N')
df_deptdescr['buyer_status'] = 'ACTIVE'
df_deptdescr = df_deptdescr.reset_index()
#(c) Include for each customer their buyer status, and the total dollar amount of the purchases 
#from XYZ using all data available for him or her.
df_customerltd = df_customer.loc[:,['acctno', 'ltd_sales']]
df_deptdescription = df_customerltd.merge(df_deptdescr, on ='acctno' )
df_deptdescriptionF = pd.concat([df_deptdescription, df_customerStatus1])
df_deptdescriptionF.fillna('N', inplace=True)

df_deptdescriptionF = df_deptdescriptionF.loc[
    :,['acctno','buyer_status','ltd_sales',
       'Appliances','Cameras & Camcorder Accessori',
       'Home Audio','Mobile Electronic Accessories',
       'Mobile Electronics','Portable Electronics',
       'Small Appliances']]
df_deptdescriptionF
```
