```python
#create a dataframe and read the excel file into the dataframe
df_2016 = pd.read_excel("2016SFOCustomerSurveyData.xlsx", sheetname=0, 
                        parse_cols = "A,CE,BN,BS,BX,BY,BZ, AK:AY, BE:BJ")
#rename the column for consistancy with other dataframes
df_2016.rename(columns={'*RESPNUM' : 'RESPNUM'}, inplace=True)
df_2016['Date'] = '2016'
```

```python
#Graded Exercise 1 - Part 2
# create a daframe with comments from the following columns
df_2016 = pd.read_excel("2016SFOCustomerSurveyData.xlsx", 
                        sheetname=0, parse_cols = "AZ:BD")
df_2016
#Used melt() to make the wide and short data into tall and narrow
df_2016_Melted = pd.melt(df_2016, var_name="Q8COMMENTS", 
                         value_name="CODE")
df_2016_Melted=df_2016_Melted['CODE'].value_counts().reset_index().rename(
    columns={'index': 'COMMENT',0: 'count'})
#Using Lambda to calculate relativefrequency
df_2016_Melted['RELATIVEFREQ']=df_2016_Melted.apply(lambda x: 
                                                      (x.CODE / df_2016_Melted
                                                       ['CODE'].sum()),
                                                      axis=1)
df_2016_Melted['COMMENT'] = df_2016_Melted['COMMENT'].astype(int)
df_2016_Melted
```

```python
#Display top 3 most comments
# create a daframe with comments from the following columns
df_2016_comments = pd.read_excel("SFO2016Comments.xlsx",
                                 sheetname=0, parse_cols = "A:B")
#df_2016_comments
df_2016_Final = pd.merge(df_2016_Melted, df_2016_comments, 
                         on='COMMENT', how='inner')
df_2016_Final.iloc[0:3]
```
